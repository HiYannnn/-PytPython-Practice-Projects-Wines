```python
import turtle as t  
from turtle import *
import random as r
import time
 
n = 100.0
 
speed("fastest")  
screensize(bg='black')  
left(90)
forward(3*n)
color("orange", "yellow") #star at top
begin_fill()
left(126)
 
for i in range(5): 
    forward(n/5)
    right(144)   
    forward(n/5)
    left(72)    
end_fill()
right(126)
 
def drawlight():
    if r.randint(0, 30) == 0: #larger range less lights
        circle(6) #size
    elif r.randint(0,30) == 1:
        color('orange')# second color
        circle(3)
    else:
        color('dark green')#empty branches
 
 
color("dark green")#branch color
backward(n*4.8)
def tree(d, s): #draw tree
    if d <= 0: return
    forward(s)
    tree(d-1, s*.8)
    right(120)
    tree(d-3, s*.5)
    drawlight()
    right(120)
    tree(d-3, s*.5)
    right(120)
    backward(s)
tree(15, n)
backward(n/2)
 
for i in range(200):#decor at bottom
    a = 200 - 400 * r.random()
    b = 10 - 20 * r.random()
    up()
    forward(b)
    left(90)
    forward(a)
    down()
    if r.randint(0, 1) == 0:
            color('tomato')
    else:
        color('wheat')
    circle(2)
    up()
    backward(a)
    right(90)
    backward(b)
 
t.color("dark red","red")#letter color
t.write("Merry Christmas",align ="center",font=("Comic Sans MS",40,"bold"))#letter,position,type,size

def drawsnowman(n,m,a,b):
    speed(100)
    t.goto(n, m)
    t.pencolor("white")
    t.pensize(2)
    t.fillcolor("white")
    t.seth(0)
    t.begin_fill()
    t.circle(a)
    t.end_fill()
    t.seth(180)
    t.begin_fill()
    t.circle(b)
    t.end_fill()
    t.pencolor("black")
    t.fillcolor("black")
    t.penup()    # right eye
    t.goto(n-a/4, m+a)
    t.seth(0)
    t.pendown()
    t.begin_fill()
    t.circle(2)
    t.end_fill()
    t.penup()    # left eye
    t.goto(n+a/4, m+a)
    t.seth(0)
    t.pendown()
    t.begin_fill()
    t.circle(2)
    t.end_fill()
    t.penup()  # mouth
    t.goto(n, m+a/2)
    t.seth(0)
    t.pendown()
    t.fd(5)
    t.penup()       # buttoms
    t.pencolor("red")
    t.fillcolor("red")
    t.goto(n, m-b/4)
    t.pendown()
    t.begin_fill()
    t.circle(2)
    t.end_fill()
    t.penup()
    t.pencolor("yellow")
    t.fillcolor("yellow")
    t.goto(n, m-b/2)
    t.pendown()
    t.begin_fill()
    t.circle(2)
    t.end_fill()
    t.penup()
    t.pencolor("orange")
    t.fillcolor("orange")
    t.goto(n, m-(3*b)/4)
    t.pendown()
    t.begin_fill()
    t.circle(2)
    t.end_fill()

drawsnowman(-200, -200, 20, 30)
drawsnowman(-250, -200, 30, 40)
 
def drawsnow():
    t.speed(0)  
    t.pensize(2)
    t.showturtle()  # Make the turtle visible
    
    for i in range(200): #num of snowflakes
        t.pencolor("white")
        t.penup()
        t.setx(r.randint(-350, 350))
        t.sety(r.randint(-100, 350)) #above the ground
        t.pendown()
        dens = 6
        snowsize = r.randint(1, 10)
        for j in range(dens):
            t.forward(int(snowsize))
            t.backward(int(snowsize))
            t.right(360 // dens)

drawsnow()
t.done()
```
